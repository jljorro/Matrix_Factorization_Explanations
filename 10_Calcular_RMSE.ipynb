```python
import pandas as pd
import numpy as np
import math
```

```python
all_ratings_df = pd.read_csv('data/experiment_data/result_clean_values.csv')
all_ratings_df.head()
```

```python
for k in range(1,11):
    all_ratings_df['error_binary_k_{}'.format(k)] = all_ratings_df.apply(lambda row: abs(row['rating_binary_k_{}'.format(k)] - row['predicted']), axis=1)
    
for k in range(1,11):
    all_ratings_df['error_q_k_{}'.format(k)] = all_ratings_df.apply(lambda row: abs(row['rating_q_k_{}'.format(k)] - row['predicted']), axis=1)
```

```python
for k in range(1, 11):
    all_ratings_df.drop(columns='rating_binary_k_{}'.format(k), inplace=True)
    all_ratings_df.drop(columns='rating_q_k_{}'.format(k), inplace=True)
```

```python
all_ratings_df.drop(columns=['predicted', 'userId', 'movieId'], inplace=True)
all_ratings_df.head()
```

```python
def num_err(values, k):
    valid_values = np.sum(~np.isnan(values))
    limit = min(k, valid_values)
            
    return limit

def sum_err(values, k):
    limit = num_err(values, k)
    sum_values = values[:limit]
    
    total = 0.0
    for v in sum_values:
        total += pow(v, 2)
        
    return total
```

```python
k_values = [1, 2, 3, 5, 10]
for k in k_values:
    all_ratings_df['sum_err_bin_k_{}'.format(k)] = all_ratings_df.apply(lambda row: sum_err(row.values[1:11], k), axis=1)
    all_ratings_df['num_err_bin_k_{}'.format(k)] = all_ratings_df.apply(lambda row: num_err(row.values[1:11], k), axis=1)
    
for k in k_values:
    all_ratings_df['sum_err_q_k_{}'.format(k)] = all_ratings_df.apply(lambda row: sum_err(row.values[11:21], k), axis=1)
    all_ratings_df['num_err_q_k_{}'.format(k)] = all_ratings_df.apply(lambda row: num_err(row.values[11:21], k), axis=1)
    
for k in range(1, 11):
    all_ratings_df.drop(columns='error_binary_k_{}'.format(k), inplace=True)
    all_ratings_df.drop(columns='error_q_k_{}'.format(k), inplace=True)
```
