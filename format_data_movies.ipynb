```python
import pandas as pd
import random
import numpy as np
```

```python
moviesDF = pd.read_csv('data/movies_data.csv')

print(moviesDF)
```

```python
def modifyCSV(row, column):
    """
        Formatear texto item1|item2|...
    """
    #print(row[0])
    
    string = row[column]
    
    string = string.replace('{', '')
    string = string.replace('}', '')
    string = string.replace("'", '')
    string = string.replace('"', '')
    string = string.replace(", ", '|')
    
    return string
```

```python
moviesDF['companies'] = moviesDF.apply (lambda row: modifyCSV(row, 'companies'), axis=1)
moviesDF['genres'] = moviesDF.apply (lambda row: modifyCSV(row, 'genres'), axis=1)
moviesDF['keywords'] = moviesDF.apply (lambda row: modifyCSV(row, 'keywords'), axis=1)
moviesDF['stars'] = moviesDF.apply (lambda row: modifyCSV(row, 'stars'), axis=1)
moviesDF['writers'] = moviesDF.apply (lambda row: modifyCSV(row, 'writers'), axis=1)

print(moviesDF)
```
